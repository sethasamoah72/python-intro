# combining two plots
* create two empty subplots above each other figsze 4x6 inch, shared y axis
* documentation: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html

```python
#solution:
fig, axes = plt.subplots(2,1, figsize=(4, 5), sharex=True)
axes[0].errorbar(x, ya, yerr=0.05, ms=8, color='dimgray', marker='s', label='gene A', capsize=4)
axes[1].errorbar(x, yb, yerr=0.05, ms=8, color='red', marker='x', label='gene B', capsize=4)
axes[1].set_xlim([0,0.11])
axes[0].set_ylim([0,1.1])
axes[1].set_ylim([0,1.1])
axes[1].set_xlabel('concentration [mol/l]', fontsize=12, weight='bold')
axes[0].set_ylabel('activity gene A', fontsize=12, weight='bold')
axes[1].set_ylabel('activity gene B', fontsize=12, weight='bold')
plt.tight_layout()
fig.subplots_adjust(hspace=0) #has to be last so tight layout does not break subplot spacing
plt.savefig('AB_subplots.pdf')
```
